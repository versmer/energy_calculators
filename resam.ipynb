```python
#reading csv and resampling using pandas
import pandas as pd
import datetime

df_supply=pd.read_csv("Lafayette_11sf_.csv", skiprows=1, low_memory=False, index_col="Date Time, GMT-05:00")
df_supply.index = pd.to_datetime(df_supply.index)


print(df_supply)
```

```python
print(df_supply.resample("H").mean())
```

```python
print(df_supply.resample("H").median())
```

```python
print(df_supply.resample("H").max())
```
